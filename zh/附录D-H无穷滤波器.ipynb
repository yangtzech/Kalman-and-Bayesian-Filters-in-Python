简单来说，$H_\infty$ (H infinity) 滤波器类似于卡尔曼滤波器，但在面对非高斯、不可预测的输入时更加健壮。

我的FilterPy库中包含了一个H-Infinity滤波器。我在下面粘贴了一些测试代码，该代码实现了上述文章中由Simon设计的滤波器。希望能对您有所帮助。

````
import numpy as np
import matplotlib.pyplot as plt

from filterpy.hinfinity import HInfinityFilter

dt = 0.1
f = HInfinityFilter(2, 1, dim_u=1, gamma=.01)

f.F = np.array([[1., dt],
                [0., 1.]])

f.H = np.array([[0., 1.]])
f.G = np.array([[dt**2 / 2, dt]]).T
````
