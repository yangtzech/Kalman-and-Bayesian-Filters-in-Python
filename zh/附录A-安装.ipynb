# 安装

本书是用Jupyter Notebook写成的，它是一种基于浏览器的交互式Python环境，混合了Python、文本和数学。我选择它是因为它具有交互特性——在交互式环境中，我发现学习卡尔曼滤波几乎是不可能的。在许多参数上很难形成直觉，直到你可以改变它们并立即看到输出结果。交互式环境还允许您进行“假设”场景的演练。 “如果我将$\mathbf{Q}$设置为零会怎样？”使用Jupyter Notebook很容易找到答案。

我选择使用 Jupyter Notebook 的另一个原因是，大多数教科书都没有透露许多细节。例如，可能会有一些伪代码旁边有一个漂亮的图表，但是该图表是由某个读者无法获得的软件生成的。我希望这本书中涉及到的所有内容都能够为您所获得。如何绘制协方差椭圆？大多数书籍都不会告诉你。使用 Jupyter Notebook，您只需要查看源代码即可。

即使您选择在线阅读本书，您也需要安装 Python 和 SciPy 栈，以便您可以编写自己的 Kalman 滤波器。有许多不同的方法可以安装这些库，我无法涵盖所有内容，但我将涵盖一些典型的场景。

安装SciPy堆栈的其他选择。您可以在这里找到说明： http://scipy.org/install.html 它可能非常麻烦，我不支持它或提供任何有关如何执行此操作的说明。

许多Linux发行版都预装了这些软件包。然而，它们通常有点过时，因此需要更新，因为本书依赖于所有最新版本。更新特定的Linux安装超出了本书的范围。Anaconda发行版的优点是它不会修改您的本地Python安装，因此您可以安装它而不破坏您的Linux发行版。有些人被这个问题绊倒了。他们安装了Anaconda，但安装的Python仍然是默认版本，然后本书的软件无法正确运行。

## SymPy

SymPy 是一个用于进行符号数学计算的 Python 包。它的全部功能超出了本书的范围，但它可以执行代数、积分和微分方程求解等操作。例如，我们使用它来计算矩阵的雅可比矩阵和期望值积分计算。

首先，一个简单的例子。我们将导入SymPy，初始化其漂亮的打印功能（它将使用LaTeX打印方程）。然后我们将声明一个符号供SymPy使用。

```
import sympy
sympy.init_printing(use_latex='mathjax')

phi, x = sympy.symbols('\phi, x')
phi
```

注意它如何使用LaTeX打印符号“phi”。现在让我们做一些数学。$\sqrt{\phi}$的导数是什么？

```
sympy.diff('sqrt(phi)')
```

我们可以分解方程

```
sympy.factor(phi**3 -phi**2 + phi - 1)
```

并且我们可以扩展它们。

```
((phi+1)*(phi-4)).expand()
```

您可以针对其变量的特定值计算方程：

```
w =x**2 -3*x +4
print(w.subs(x, 4))
print(w.subs(x, 12))
```

您也可以使用字符串表示使用未定义符号的方程：

```
x = sympy.expand('(t+1)*2')
x
```

现在让我们使用SymPy计算矩阵的Jacobian。给定函数

$$h=\sqrt{(x^2 + z^2)}$$

求相对于 x、y 和 z 的雅可比矩阵。

```
x, y, z = sympy.symbols('x y z')

H = sympy.Matrix([sympy.sqrt(x**2 + z**2)])

state = sympy.Matrix([x, y, z])
H.jacobian(state)
```

现在让我们计算离散过程噪声矩阵 $\mathbf Q$，给定连续过程噪声矩阵为
$$\mathbf Q = \Phi_s \begin{bmatrix}0&0&0\\0&0&0\\0&0&1\end{bmatrix}$$

积分是

$$\mathbf Q = \int_0^{\Delta t} \mathbf F(t)\mathbf Q\mathbf F^T(t)\, dt$$

其中
$$\mathbf F(\Delta t) = \begin{bmatrix}1 & \Delta t & {\Delta t}^2/2 \\ 0 & 1 & \Delta t\\ 0& 0& 1\end{bmatrix}$$

```
dt = sympy.symbols('\Delta{t}')
F_k = sympy.Matrix([[1, dt, dt**2/2],
                    [0,  1,      dt],
                    [0,  0,       1]])
Q = sympy.Matrix([[0,0,0],
                  [0,0,0],
                  [0,0,1]])

sympy.integrate(F_k*Q*F_k.T,(dt, 0, dt))
```
