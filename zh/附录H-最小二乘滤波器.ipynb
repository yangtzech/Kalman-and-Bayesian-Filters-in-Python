在本章的开头，我使用了`numpy.polyfit()`函数来将一条直线拟合到重量测量数据上。它使用最小二乘法将一个n次多项式拟合到数据中。这与g-h滤波器有何不同？

嗯，这取决于情况。我们最终会学到，在某些条件下，卡尔曼滤波器从最小二乘拟合的角度来看是最优的。然而，`polyfit()`函数将一个多项式拟合到数据上，而不是一个任意的曲线，通过最小化这个公式的值来实现：

$$E = \sum_{j=0}^k |p(x_j) - y_j|^2$$

我曾经假定自己的体重每天增加1磅，因此当我尝试拟合$n=1$的多项式（即一条直线）时，结果非常接近实际体重增加量。但是，当然，并不是每个人都一直只增重或减重。我们会波动。对于更长的数据序列使用'polyfit（）'将产生较差的结果。相比之下，g-h滤波器对速率的变化做出反应 - $h$项控制滤波器对这些变化的反应速度。如果我们增重，保持稳定一段时间，然后减重，滤波器将自动跟踪该变化。除非增重和减重可以很好地表示为多项式，否则'polyfit（）'将无法做到这一点。

这种滤波器的另一个优点是，即使数据符合 *n* 次多项式，它也是 *递归* 的。也就是说，我们可以在不知道上一个时间段的估计值和速率的情况下计算出本时间段的估计值。相比之下，如果你深入了解 `polyfit()` 的实现，你会发现它需要所有数据才能生成答案。因此，像 `polyfit()` 这样的算法不适合实时数据过滤。在 60 年代开发 Kalman 滤波器时，计算机非常缓慢，内存极其有限。它们完全无法存储例如来自飞机惯性导航系统的数千个读数，也无法在需要提供准确和最新导航信息的短时间内处理所有这些数据。

到了20世纪中期，各种最小二乘估计的形式被用于这种类型的滤波器。例如，NASA的阿波罗计划有一个地面网络用于跟踪指令服务模块（CSM）和登月舱（LM）。他们进行了多分钟的测量，将数据分批处理，并慢慢计算出答案。1960年，NASA艾姆斯的斯坦利·施密特（Stanley Schmidt）认识到了鲁道夫·卡尔曼（Rudolf Kalman）的开创性论文的实用性，并邀请他来到艾姆斯。施密特将卡尔曼的工作应用于CSM和LM上的机载导航系统，并称之为“卡尔曼滤波器”。[1]不久之后，世界转向了这种更快、递归的滤波器。

卡尔曼滤波器只需要存储最后一次估计和一些相关参数，并且仅需要相对较少的计算即可生成下一个估计。现在我们拥有如此多的内存和处理能力，这种优势已经不太重要了，但在当时，卡尔曼滤波器不仅因为数学特性而是一项重大突破，而是因为它（勉强）可以在当时的硬件上运行。
