# 使用泰勒级数进行线性化

泰勒级数将函数表示为无穷项之和。这些项是线性的，即使对于非线性函数也是如此，因此我们可以使用线性代数来表示任意非线性函数。这样做的代价是，除非我们使用无穷多项，否则计算出的值将是近似值而非精确值。

实数或复数函数 f(x) 在 x=a 处的泰勒级数是如下无穷级数

$$f(x) = f(a) + f'(a)(x-a) + rac{f''(a)}{2!}(x-a)^2 + \, ...\,  + rac{f^{(n)}(a)}{n!}(x-a)^n + \, ...$$

其中 $f^{n}$ 是 f 的 n 阶导数。为了计算 $f(x)=sin(x)$ 在 $x=0$ 处的泰勒级数，让我们首先求出 f 的各项值。

$$egin{aligned}
f^{0}(x) &= sin(x) ,\ \  &f^{0}(0) &= 0 \
f^{1}(x) &= cos(x),\ \  &f^{1}(0) &= 1 \
f^{2}(x) &= -sin(x),\ \  &f^{2}(0) &= 0 \
f^{3}(x) &= -cos(x),\ \  &f^{3}(0) &= -1 \
f^{4}(x) &= sin(x),\ \  &f^{4}(0) &= 0 \
f^{5}(x) &= cos(x),\ \  &f^{5}(0) &= 1
\end{aligned}
$$

现在我们可以将这些值代入方程。

$$\sin(x) = rac{0}{0!}(x)^0 + rac{1}{1!}(x)^1 + rac{0}{2!}(x)^2 + rac{-1}{3!}(x)^3 + rac{0}{4!}(x)^4 + rac{1}{5!}(x)^5 + ... $$

让我们用一些代码来验证：

```python
import numpy as np

x = .3
estimate = x - x**3/6 + x**5/120
exact = np.sin(.3)

print('sin(.3) 的近似值为', estimate)
print('sin(.3) 的精确值为', exact)
```

仅用三项就能得到这样的结果，已经相当不错了。如果你有兴趣，可以尝试将其编写为一个 Python 函数，来计算任意多项的级数。
